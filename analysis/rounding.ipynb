```python
import numpy as np
import matplotlib.pyplot as plt

sigma_vals = 10**np.linspace(-3.,3.,1000)
res1 = []
res2 = []
for sigma in sigma_vals:
    X = np.random.default_rng(seed=0).normal(0,sigma,(1000000,))
    res1.append(np.mean(np.abs(X)))    
    res2.append(np.mean(np.abs(np.round(X))))
plt.figure()
plt.semilogx(sigma_vals, np.array(res1)-np.array(res2))
plt.xlabel('sigma')
#plt.legend(('Unrounded','Rounded'))
plt.ylabel('Gain from rounding')
plt.grid(True)
```
